# Part 1: authority.xlsx

```python
f = '.../authority.xlsx'
# Replace 'YOUR_FILE_PATH' as the path to the authority.xlsx file

# Do not delete the code below
authority = pd.read_excel(f)
print(authority.shape)
print(authority.dtypes)
authority.head()
```

```python
# 1. How many unique author IDs are in the file?
19473
# Your code for anwsering the question:
print(authority['Author ID'].nunique())
```

```python
# 2. How many authors are there in each paper?

# Your code for anwsering the question:
# Your code for step 1:
author_counts_1 = authority.groupby(['PMID'])[['Author ID']].count()

# Your code for step 2:
author_counts_1 = author_counts_1.reset_index()
# Your code for step 3:
author_counts_1 = author_counts_1.rename(columns={'Author ID':'author_counts_authority'})

# Your code for step 4:
author_counts_1 = author_counts_1.sort_values(by=['author_counts_authority'], ascending=False)

# Your code for step 5:
author_counts_1.head()
```

```python
# 3.How many papers have more than five authors?
1156
# Your code for anwsering the question:
author_counts_2 = author_counts_1.loc[author_counts_1['author_counts_authority']>5]
author_counts_2.shape
```

# Part 2: icite.xlsx

```python
f = '...../icite.xlsx'
# Replace 'YOUR_FILE_PATH' as the path to the icite.xlsx file

# Do not delete the code below
icite = pd.read_excel(f)
icite = icite.rename(columns={'pmid':'PMID'})
print(icite.shape)
print(icite.dtypes)
icite.head()
```

```python
# 1.How many papers were published between 2000 and 2010?
1310
# Your code for anwsering the question:
# Your code for step 1:
filter_1 = icite['year'] >= 2000
filter_2 = icite['year'] <= 2010
in_year_range = icite.loc[(filter_1)&(filter_2)]

# Your code for step 2:
in_year_range.shape
```

```python
# 2. How many papers are research articles? How many are not?
# 4286 are research articles. 714 are not.
# Your code for anwsering the question:
grouping_result = icite.groupby(['is_research_article'])[['PMID']].count()
grouping_result
```

```python
# 3. How many papers have a DOI? 3556

# Your code for anwsering the question:
# Your code for step 1:
papers_with_doi = icite[~pd.isnull(icite['doi'])]
# Your code for step 2:
papers_with_doi.shape
```

```python
# 4. What are the minimum, maximum, mean, and median of the citation counts?
# minimum: 0.000000, maximum: 1923.000000, mean: 23.713400, median = 7.000000
# Your code for anwsering the question:
icite.describe()
```

```python
# 5. How many authors are there in each paper?

# Your code for anwsering the question:
icite['authors_splitted'] = icite['authors'].str.split('|')
icite['author_counts_icite'] = icite['authors_splitted'].str.len()
icite.head()
```

# Part 3: Merging dataframes

```python
# 0. Prepare a table for merging

# Your code for anwsering the question:
selected_cols =['PMID', 'author_counts_icite']
icite_selected = icite[selected_cols]
```

```python
# 1. Merge the prepared table with the author_counts_1 dataframe

# Your code for anwsering the question:
# Your code for step 1:
merged_df = icite_selected.merge(author_counts_1)

# Your code for step 2:
merged_df = icite_selected.merge(author_counts_1.rename(columns={'author_counts_authority':'author_counts_authority_lookup'}))

# Your code for step 3:
merged_df.shape

# Your code for step 4:
merged_df
```
